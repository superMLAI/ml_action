```python
import numpy as np
import operator
#P19  knn分类器代码
def classify0(inX,dataSet,labels,k):     #labels标签向量的元素数目和dataSet训练样本矩阵的行数相同
    dataSetSize=dataSet.shape[0]
    diffMat=np.tile(inX,(dataSetSize,1))-dataSet   #使用欧氏距离公式，计算两个向量的距离
    sqDiffMat=diffMat**2
    sqDistances=sqDiffMat.sum(axis=1)
    distances=sqDistances**0.5
    sortedDistIndicies=distances.argsort()    #返回数值从小到大排序后对应原位置的索引...相当于告诉你第一名去哪找，第二名去哪找
    classCount={}
    for i in range(k):    #选择距离最近的k个点
        voteIlabel=labels[sortedDistIndicies[i]]
        classCount[voteIlabel]=classCount.get(voteIlabel,0)+1    #没有当前类别就初始化为0，有就在对应类别计数加1
    sortedClassCount=sorted(classCount.items(),key=operator.itemgetter(1),reverse=True)#key:value,所以operator.itemgetter(1)表示按value排序
    return sortedClassCount[0][0]
```

```python
#如何在二进制存储的图像数据上使用knn
#为了使用前面例子的分类器classify0，必须将32*32的二进制图像矩阵格式化为一个1*1024的向量
def img2vector(filename):
    returnVect=np.zeros((1,1024))
    fr=open(filename)
    for i in range(32):
        lineStr=fr.readline()
        for j in range(32):
            returnVect[0,32*i+j]=int(lineStr[j])     #将32*32矩阵中的每个二进制数据(0或1)对应到1*1024向量的第一行的第？列
    return returnVect
```

```python
from os import listdir          #列出给定目录下的文件名
def handwritingClassTest():
    hwLabels=[]                 #训练样本的类别标签
    trainingFileList=listdir('Reference code/trainingDigits')    #取出所有训练样本的文件名
    m=len(trainingFileList)     #统计训练样本数量
    trainingMat=np.zeros((m,1024))   #初始化用于测试的，由m个训练样本转化成的m*1024训练矩阵
    for i in range(m):          #逐个操作训练样本文件，m个，m次循环
        fileNameStr=trainingFileList[i]        #取出当前循环的文件名字符串
        classNumStr=int(fileNameStr.split('_')[0])   #用'_'分割文件名，分割后第一个元素就是当前样本的类别标签，转化为整型
        hwLabels.append(classNumStr)           #将生成的样本类别标签依次添加到用于测试的标签列表中
        trainingMat[i,:]=img2vector('Reference code/trainingDigits/%s'%fileNameStr) #将当前循环中文件生成的1*1024向量，输入训练矩阵的对应行
    testFileList=listdir('Reference code/testDigits')   #读取测试样本文件名
    errorCount=0
    mTest=len(testFileList)
    for i in range(mTest):
        fileNameStr=testFileList[i]
        fileStr=fileNameStr.split('.')[0]
        classNumStr=int(fileStr.split('_')[0])
        vectorUnderTest=img2vector('Reference code/testDigits/%s'% fileNameStr)
        classifierResult=classify0(vectorUnderTest,trainingMat,hwLabels,3)
        print('the classifier came back with: %d,the real answer is: %d'%(classifierResult,classNumStr))
        if (classifierResult!=classNumStr):errorCount+=1
    print('\nthe total number of errors is: %d'%errorCount)
    print('\nthe total error rate is: %f'%(errorCount/mTest))
```

```python
handwritingClassTest()
```
